# Question 2

# Question 3
Which artist is created the most artwork by total surface area?

For this question, I need to calculate surface of each artwork. However, it's challenging as artworks can have various shapes (sculpture, paintings, objects...)

I need to take hypothesis to estimate sufaces :

- when there is a diameter, the artwork can be a circle (other dimensions columns null) OR a 3-dimension object (height column not null). To simplify my analysis, I will assume that those shapes are ovale.

- when there isn't diameter, I'm gonna assume (to simplify my analysis) that artworks are rectangular or square (most common shapes of artworks).


```python
Question3 = pd.read_sql(""" SELECT Name, CASE 
                        WHEN "Diameter (cm)" NOT NULL 
                        AND "Height (cm)" IS NULL
                            THEN (3.14*(("Diameter (cm)"/2)*("Diameter (cm)"/2)))
                        WHEN "Diameter (cm)" NOT NULL 
                        AND "Height (cm)" NOT NULL
                            THEN (3.14*(("Diameter (cm)"/2)*("Height (cm)"/2)))
                        WHEN "Diameter (cm)" IS NULL 
                        AND "Height (cm)" IS NULL AND "Length (cm)" NOT NULL AND "Width (cm)" NOT NULL
                            THEN ("Length (cm)"*"Width (cm)")
                        WHEN "Diameter (cm)" IS NULL
                        AND "Length (cm)" IS NULL AND "Width (cm)" NOT NULL AND "Height (cm)" NOT NULL
                            THEN ("Width (cm)"*"Height (cm)")
                        WHEN "Diameter (cm)" IS NULL
                        AND "Width (cm)" IS NULL AND "Length (cm)" NOT NULL AND "Height (cm)" NOT NULL
                            THEN ("Length (cm)"*"Height (cm)")
                        ELSE 0 
                        END AS Surface
                        FROM artworks
                        ORDER BY Surface DESC
                        LIMIT 1;""", conn)
Question3
```

# Question 4
Did any artists have artwork acquired during their lifetime?

I need to join the two tables and:

- for dead artists, verify if the year of AcquisitionDate of each artwork is between the Birth_Year and the Death_Year of the artist

- for artists alive (we assume that those who don't have a Death_Year are still alive), add them to the analysis

Note that from my exploratory analysis, I know that some artists don't have a Birth_Year and should be removed from this analysis. 

```python
# DEAD artists who have artwork acquired during their lifetime
Question4a = pd.read_sql("""SELECT artt.Name
                        FROM artists artt
                        JOIN artworks artw
                        ON artt.Artist_ID = artw.ArtistID
                        WHERE strftime('%Y',artw.AcquisitionDate) <= artt.Death_Year
                        GROUP BY artt.Name
                        ;""", conn)
Question4a
```

```python
# Number of dead artists who have artwork acquired during their lifetime
len(Question4a)
```

```python
# Add artists alive (death year is null AND birth year is not null - as some artists don't have a birth year)
Question4b = pd.read_sql("""SELECT artt.Name
                        FROM artists artt
                        JOIN artworks artw
                        ON artt.Artist_ID = artw.ArtistID
                        WHERE (strftime('%Y',artw.AcquisitionDate) <= artt.Death_Year AND artt.Birth_Year IS NOT NULL)
                        OR (artt.Death_Year IS NULL AND artt.Birth_Year IS NOT NULL)
                        GROUP BY artt.Name
                        ;""", conn)
Question4b
```

```python
# Number of artists (both dead and alive) who have/had artwork acquired during their lifetime
len(Question4b)
```

# Question 6
Please group the artworks into as many clusters as you feel is appropriate, using attributes from both the artist and artworks tables, and assign each artwork to this new cluster.

#### Grouping by classification/department

```python
#First, taking the artwork table, we can cluster artworks by department and/or classification. 
#The analysis will be exhaustive as there is no missing value in those two columns.
Question6a = pd.read_sql("""SELECT Classification, Department,
                        COUNT(ArtworkID) as Num_Artworks,
                        100*(COUNT(ArtworkID)/cast((SELECT COUNT(ArtworkID) FROM artworks) AS float)) as PerTotal
                        FROM artworks 
                        GROUP BY Classification
                        ORDER BY Num_Artworks Desc
                        limit 10
                    """, conn)
Question6a
```

#### Grouping by gender

```python
# Grouping artworks by artists' gender (however, note that the analysis is not exhaustive as we have missing values in our dataset).
# For Male artists, there are 2 writings so we need to add a condition to group them in the same row.
Question6b = pd.read_sql("""SELECT CASE WHEN artt.Gender='male' THEN 'Male' ELSE artt.Gender END AS Gender,
                        COUNT(artw.ArtworkID) AS Num_Artworks,
                        (COUNT(artw.ArtworkID)/CAST((SELECT COUNT(ArtworkID) FROM artworks) AS float)*100) AS Per_Total
                        FROM artists artt
                        INNER JOIN artworks artw ON artt.Artist_ID=artw.ArtistID
                        GROUP BY 1
                        ORDER BY Num_Artworks Desc
                    """, conn)
Question6b
```

#### Grouping by nationality

```python
# Grouping artworks by artists' nationality (being however that the analysis is not exhaustive as we have missing values )
Question6c = pd.read_sql("""SELECT artt.Nationality,
                        COUNT(artw.ArtworkID) AS Num_Artworks,
                        (COUNT(artw.ArtworkID)/CAST((SELECT COUNT(ArtworkID) FROM artworks) AS float)*100) AS Per_Total
                        FROM artists artt
                        INNER JOIN artworks artw ON artt.Artist_ID=artw.ArtistID
                        GROUP BY 1
                        ORDER BY Num_Artworks Desc
                        LIMIT 4
                    """, conn)
Question6c
```

#### Grouping by gender & nationality

```python
# Grouping artworks by gender and nationality
Question6d = pd.read_sql("""SELECT CASE WHEN artt.Gender='male' THEN 'Male' ELSE artt.Gender END AS Gender,
                        artt.Nationality,
                        COUNT(artw.ArtworkID) AS Num_Artworks,
                        (COUNT(artw.ArtworkID)/CAST((SELECT COUNT(ArtworkID) FROM artworks) AS float)*100) AS Per_Total
                        FROM artists artt
                        INNER JOIN artworks artw ON artt.Artist_ID=artw.ArtistID
                        GROUP BY 1, 2
                        ORDER BY Num_Artworks Desc
                        LIMIT 10
                    """, conn)
Question6d
```
